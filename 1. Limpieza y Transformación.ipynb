## 5.2. Nulos de puntajes

```python
df[df.puntaje_global.isnull()]
```

Los registros que tienen cero resultados publicados es normal que tengan nulos en los puntajes, por ellos se decide eliminar estos registros.

```python
df.drop(df[df.publicados==0].index, axis=0, inplace=True)
```

En los otros dos registros con nulos, el problema es que no tienen resultados en la prueba de inglés. Los nulos o ceros de la prueba de inglés se van a reemplazar con la mediana del establecimiento en los otros semestres con registros válidos.

```python
#Observar los registros anteriores para los establecimientos con nulos en puntajes
cols = ['puntaje_ingles', 'ponderado_ingles', 'punt_ingles_med', 'punt_ingles_sector', 'punt_ingles_comu']
establecimientos = df[df.puntaje_global.isnull()].establecimiento
df[df.establecimiento.isin(establecimientos)][['establecimiento']+cols]
```

El establecimiento 'c.e. de cómputo empresarial para el trabajo - cindecóputo' solo tiene un registro, por lo que no hay manera de obtener información sobre su desempeño en otros semestres. Se decide eliminar este registro.

```python
#Eliminar el registros 323
df.drop(323, axis=0, inplace=True)
```

```python
#Obtener medianas de los puntajes de inglés en otros semestres
medianas = df[df.establecimiento.isin(establecimientos)][cols].drop(322, axis=0).median()

#Reemplazar los nulos de los puntajes de inglés por las medianas
df.loc[322, cols] = medianas.values

#Reemplazar los puntajes globales por la suma de los puntajes en las 5 asignaturas
cols2 = ['puntaje_global', 'ponderado_global', 'punt_global_med', 'punt_global_sector', 'punt_global_comu']
for col in cols2:
    pos = df.loc[[322]].columns.get_loc(col)
    df.loc[322, col] = df.iloc[322, pos+1]+df.iloc[322, pos+2]+df.iloc[322, pos+3]+df.iloc[322, pos+4]+df.iloc[322, pos+5]
```

## 5.3. Nulos de matriculados

Para estos también se usará la técnica de mediana por grupo, es decir, por establecimiento en los demás semestres con no nulos.

```python
#Obtener los establecimientos con nulos en la columna matriculados
establecimientos = df[df.matriculados.isnull()].establecimiento.unique()

#Obtener las medianas de matriculados de cada establecimiento en los semestres sin nulos
medianas = [df[df.establecimiento == estab].matriculados.median() for estab in establecimientos]

dic_reemplazo = {establecimientos[i] : medianas[i] for i in range(len(medianas))}

#Obtener indices de fila de los registros con nulos en matriculados
indices_nulos = df[df.matriculados.isnull()].index
#Reemplazar los nulos
df.loc[indices_nulos, 'matriculados'] = df[df.matriculados.isnull()].establecimiento.map(dic_reemplazo)
```

## 5.4. Revisar nuevamente los nulos

```python
nulos = df.isnull().sum()
nulos[nulos != 0]
```

# 7. Variables numéricas

### Obervar si hay valores negativos

```python
df_num = df.select_dtypes(exclude=['object', 'datetime'])
df_num_neg = df_num[df_num<0].sum()
df_num_neg[df_num_neg<0]
```

```python
df[df.v4_6<0]
```

Un valor negativo en cualquiera de las columnas que representan la morosidad de la cartera de préstamos es inusual. Por lo general, se espera que los valores en estas columnas sean números enteros positivos o cero, que reflejan la cantidad de préstamos o el saldo de préstamos en morosidad durante un período específico. Un valor negativo podría ser el resultado de errores en la recopilación, el procesamiento o la entrada de datos.

Corregir valores negativos

```python
indices = df[df.v4_6<0].index
df.loc[indices, 'v4_6'] = df.loc[indices, 'v4_6']*-1
```

Como la columna vigente depende de los abonos a las cuentas (saldo a la fecha - abonos), también debe corregirse esta columna.

```python
abonos = df.iloc[indices, 8:21].sum(axis=1).values
df.iloc[indices, 6:8].vigente = df.iloc[indices, 6:8].saldo_a_la_fecha.values - abonos
```

```python
df_num = df.select_dtypes(exclude=['object', 'datetime'])
df_num_neg = df_num[df_num<0].sum()
df_num_neg[df_num_neg<0]
```

### Observar si hay algún valor de cero en la columna saldo total a la fecha.

```python
df[df.saldo_a_la_fecha<=0]
```

Valores de cero en esta columna se pueden interpretar como nulos. Podemos imputarlos con la suma de las columnas de saldo por riesgo (riesgo_a_s, riesgo_b_s, ...), ya que estas columnas representan la parte del saldo total que pertenece a cada riesgo. Esto solo en caso de que haya valores diferentes de cero en las columnas de saldo por riesgo y en las columnas de abonos (v1_2, v_2_3, ...).

Primero observemos si hay registros con cero saldo a la fecha y valores diferentes de cero en las columnas de abonos.

```python
#Si la suma es cero es porque todas las filas son cero.
df[df.saldo_a_la_fecha<=0].iloc[:,6:21].sum()
```

Como se evidencia, no hay ninguno, por tanto, se eliminan todos los registros con ceros en saldo a la fecha.

```python
#Filtrar las sumas iguales a cero y obtener índices. Si la suma es cero es porque todas las columnas son cero.
indices_ceros = df[df.saldo_a_la_fecha<=0].index

#Eliminar los registros
df.drop(indices_ceros, axis=0, inplace=True)
```
